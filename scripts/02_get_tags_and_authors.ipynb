### GET PRIX

```python
df_prix = pd.read_csv(r'../metadata/csv_prix_all.csv', encoding='latin')#csv_prix_all/customed
```

```python
len(df_prix)
```

```python
liste_net = [unidecode(elem).lower().replace('-',' ').replace("'",' ') for elem in df_prix.Roman]
```

```python
def get_roman(list_roman):
    list_titre = []
    for elem in list_roman:
        titre = elem.split('_')[2]
        list_titre.append(' '.join(titre.lower().split('-')))
    return list_titre
```

```python
list_romans_chapitre = get_roman(list(df_main.index))
```

```python
def get_award(list_corpus, list_prix):
    list_award = []
    for elem in list_corpus:
        if elem in list_prix:
            list_award.append('awarded')
        else:
            list_award.append('not_awarded')
    return list_award
```

```python
list_award = get_award(list_romans_chapitre, liste_net)
```

```python
df_res = df_main.iloc[:, 0:4157]# pour motifs
```

```python
df_res["AWARD"] = list_award
#df_main["AWARD"] = list_award
```

### GET AUTHORS

```python
def get_df_with_authors_dates(df):
    df_res = df.copy()
    list_author, list_date = [], []
    for elem in df_res.index:
        list_author.append(elem.split('_')[1])
        list_date.append(int(elem.split('_')[0]))
    print(len(set(list_author)))
    df_res['auteur'] = list_author
    df_res['date'] = list_date
    return df_res
```

```python
df_award_authors = get_df_with_authors_dates(df_res)#df_main
```

```python
df_final = df_award_authors[df_award_authors['date']>=1900]
```

```python
def post_process(df, len_max):
    len_ngrams = 1
    conteur_temp = 0
    good_cols = []
    for col in list(df.columns):
        if len(col.split('_')) == len_ngrams and conteur_temp < len_max:
            conteur_temp+=1
            good_cols.append(col)
        elif conteur_temp >= len_max:
            len_ngrams+=1
            conteur_temp=0

    return pd.DataFrame(df, columns=good_cols)
```
